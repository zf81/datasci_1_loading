This dataset provides provisional counts for drug overdose deaths based on a current mortality data in the National Vital Statistics System. National provisional counts include deaths occurring within the 50 states and the District of Columbia and may not include all deaths that occurred during a given time period. Provisional counts are often incomplete and causes of death may be pending investigation which can result in an underestimate of counts.

Description of features:


*   State where death occurred listed by initials
*   Year in which death occurred
*   Month in which death occurred
*   Period- provisional drug overdose deaths for the current 12 month-ending period
*   Indicator- Drug type/ drug class such as heroin, opioids, psychostimulants, synthetic opioids
*   Percent Complete- Percent of drug overdose deaths which have completed investigation
*   Percent Pending Investigation- Percent of drug overdose deaths which are pending investigation
*   State where death occurred listed by name

```python
dataset4 = pd.read_json("/content/JSON/Covid Tracking Historic US values Dataset.json")
dataset4.head(10)
```

```python
dataset4.columns
```

This dataset displays COVID tracking in the United States. Of note, the COVID Tracking Project has ended all data collection as of March 7, 2021

Description of features:
*   date- Date on which data was collected by The COVID Tracking Project
*   states- The number of states and territories included in the US dataset for this day
*   positive- Cases (confirmed plus probable)
*   negative- Negative PCR tests (people)
*   pending- Total number of viral tests that have not been completed as reported by the state or territory.
*   hospitalizedCurrently- Individuals who are currently hospitalized with COVID-19
*   hospitalizedCumulative- Total number of individuals who have ever been hospitalized with COVID-19
*   inIcuCurrently- Individuals who are currently hospitalized in the Intensive Care Unit with COVID-19.
*   inIcuCumulative- Total number of individuals who have ever been hospitalized in the Intensive Care Unit with COVID-19
*   onVentilatorCurrently- Individuals who are currently hospitalized under advanced ventilation with COVID-19
*   onVentilatorCumulative- Total number of individuals who have ever been hospitalized under advanced ventilation with COVID-19
*   death- Total fatalities with confirmed OR probable COVID-19 case diagnosis
*   recovered- Total number of people that are identified as recovered from COVID-19
*   deathIncrease- Daily increase in death, calculated from the previous days value
*   hospitalizedIncrease- Daily increase in hospitalizedCumulative, calculated from the previous days value
*   negativeIncrease- Increase in negative computed by subtracting the value of negative for the previous day from the value for negative from the current day
*   positiveIncrease- The daily increase in API field positive, which measures Cases (confirmed plus probable) calculated based on the previous days value
*   totalTestResultsIncrease- Daily increase in totalTestResults, calculated from the previous days value
